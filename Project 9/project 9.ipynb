```python
plt.figure(figsize=(8,5))
data['fuel_type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['city_mpg'], bins=20, color='green', edgecolor='black')
plt.title('City MPG')
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['city_mpg'], data['highway_mpg'], alpha=0.5, color='red')
plt.title('City MPG vs Highway MPG')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.grid(True, alpha=0.3)
plt.show()
```
